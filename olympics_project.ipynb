##### Q. Calculate the medals per each country for the best male and females athletes and visualize the results

```python
data.pivot_table(index='Country',columns='Medal',values='season',aggfunc='count').fillna(0)
```

```python
data_piv=data.pivot_table(index='Country',columns='Medal',values='season',aggfunc='count').fillna(0)
data_piv['sum_Medal']=data_piv.Bronze + data_piv.Gold  +data_piv.Silver

plt.figure(figsize=(30,20))
data_piv.sort_values(by='sum_Medal' ,ascending=False).drop('sum_Medal',axis=1).head(10).plot.barh(width=0.8)
fig = plt.gcf()
fig.set_size_inches(12,12)
data_piv.sort_values(by='sum_Medal' ,ascending=False)
```

```python
data_men=data[data['Gender']=='Men']

data_men=data_men.pivot_table(index='Athlete',columns='Medal',values='season',aggfunc='count').fillna(0)
data_men['sum_Medal']=data_men.Bronze + data_men.Gold  +data_men.Silver

plt.figure(figsize=(30,20))
data_men.sort_values(by='sum_Medal' ,ascending=False).drop('sum_Medal',axis=1).head(10).plot.barh(width=0.8)
fig = plt.gcf()
fig.set_size_inches(12,12)
data_men.sort_values(by='sum_Medal' ,ascending=False)
```

```python
data_women=data[data['Gender']=='Women']

data_women=data_women.pivot_table(index='Athlete',columns='Medal',values='season',aggfunc='count').fillna(0)
data_women['sum_Medal']=data_women.Bronze + data_women.Gold  +data_women.Silver

plt.figure(figsize=(30,20))
data_women.sort_values(by='sum_Medal' ,ascending=False).drop('sum_Medal',axis=1).head(10).plot.barh(width=0.8)
fig = plt.gcf()
fig.set_size_inches(12,12)
data_women.sort_values(by='sum_Medal' ,ascending=False)
```
